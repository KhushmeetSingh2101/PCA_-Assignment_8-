```python
Wine= wine.iloc[:,1:]
Wine
```

```python
wine_ary=Wine.values
wine_ary
```

```python
wine_norm=scale(wine_ary)
wine_norm
```

```python
# Applying PCA Fit Transform to dataset
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
pca_values
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
plt.plot(Var)
```

```python
# Final Dataframe
final_df=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PCA1','PCA2','PCA3'])],axis=1)
final_df
```

```python
fig=plt.figure(figsize=(11,9))
sns.scatterplot(data=final_df)
```

```python
sns.scatterplot(data=final_df, x='PCA1', y='PCA2', hue='Type')
```

#Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

HYPERPARAMETER --> COMPLETE

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wine1=wine.copy()
wine1['clustersid']=hclusters.labels_
wine1
```

#KMEANS

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
clusters3=KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
wine2=wine.copy()
wine2['clusters3id']=clusters3.labels_
wine2
```

```python
wine2['clusters3id'].value_counts()
```
